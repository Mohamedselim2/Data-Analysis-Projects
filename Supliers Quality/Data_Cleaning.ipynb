```python
import pandas as pd

# Load the Excel file
file_path = r"C:\\Users\\Youssef\\Downloads\\Defects_updated5.xlsx"  # Update this path if necessary
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Identify duplicates in the 'Defect' column
duplicates_info = df[df.duplicated(subset=['Defect'], keep=False)]

# Store the IDs of the removed duplicates
removed_ids = []

# Iterate through the duplicates and prepare the info
for defect in duplicates_info['Defect'].unique():
    # Get the IDs of the duplicates for each defect
    ids = duplicates_info[duplicates_info['Defect'] == defect]['Defect ID'].tolist()
    # Keep the first ID and add the rest to removed_ids
    if len(ids) > 1:
        removed_ids.extend(ids[1:])  # Add all but the first ID to the removed list

# Remove duplicates from the 'Defect' column
df_unique = df.drop_duplicates(subset=['Defect'])

# Save the updated file without duplicates
output_file_path = r"C:\\Users\\Youssef\\Downloads\\Defects_without_duplicates.xlsx"
df_unique.to_excel(output_file_path, index=False)

# Print the IDs that were removed
if removed_ids:
    print("Removed the following IDs due to duplicates:")
    for removed_id in removed_ids:
        print(removed_id)
else:
    print("No duplicates found.")

print(f"File saved without duplicates to: {output_file_path}")
```

```python
import pandas as pd
from scipy.stats.mstats import winsorize

# Load the Excel file
df = pd.read_excel("C:\\Users\\Youssef\\Downloads\\Supliers_dataset.xlsx")

# Calculate the first and third quartiles
Q1 = df['Defect Qty'].quantile(0.25)
Q3 = df['Defect Qty'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Print the bounds
print(f"Lower bound for Winsorizing: {lower_bound}")
print(f"Upper bound for Winsorizing: {upper_bound}")

# Count original values
original_count = df['Downtime min'].count()

# Winsorize values outside the bounds
df.loc[df['Defect Qty'] < lower_bound, 'Defect Qty'] = lower_bound
df.loc[df['Defect Qty'] > upper_bound, 'Defect Qty'] = upper_bound

# Count how many values were adjusted
adjusted_count = df.loc[(df['Defect Qty'] < lower_bound) | (df['Defect Qty'] > upper_bound), 'Defect Qty'].count()

# Print summary of adjustments
print(f"Handled Defect Qty outliers with Winsorizing.")


# Save the modified data to a new Excel file
df.to_excel("C:\\Users\\Youssef\\Downloads\\Supliers_dataset_winsorized.xlsx", index=False)

print("Data saved to 'Supliers_dataset_winsorized_qty.xlsx'.")
```
